```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
%matplotlib inline
```

```python
Data = pd.read_csv("customerdata4.csv")
Data.head()
```

```python
Data.describe()
```

```python
sns.boxplot(data=Data, x="Age", y ="Gender")
```

```python
sns.boxplot(data=Data, x="Annual Income (10kkr)", y ="Gender")
```

```python
sns.boxplot(data=Data, x="Spending Score (1-100)", y ="Gender")
```

```python
sns.scatterplot(data=Data, x="Age", y ="Spending Score (1-100)")
```

```python
sns.scatterplot(data=Data, x="Annual Income (10kkr)", y ="Spending Score (1-100)")

points = Data[["Annual Income (10kkr)", "Spending Score (1-100)"]]

print(points)
```

```python
k = 3
kmeans = KMeans(n_clusters=k).fit(points)
centroids = kmeans.cluster_centers_
print(centroids)

plt.scatter(x = points["Annual Income (10kkr)"], y = points["Spending Score (1-100)"], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=40,marker='x')
plt.show()
```
