**Task code written below:**

````python
# write your Python code for Key here:
def getKey(file):
    myFile = open(file, "r")
    myKey = myFile.read()
    
    intHexPairs = []
    
    for i in range(0, len(myKey), 2):
        hexPair = myKey[i:i+2] # get two characters at a time
        intHexPair = int(hexPair, 16) # convert to integer
        intHexPairs.append(intHexPair) # add a integer to list
        
    return intHexPairs
    
key = getKey("testKey")  
print(key)
````

**Task code written below:**

````python
# shiftRows() here:

def rotate(word, n): 
    return word[n:] + word[:n] # rotate a word by n bits


def shiftRows(block):
    
    for i in range(4):
    
        block[i::4] = rotate(block[i::4],i) # rotate each row by i bits
    
    return block  

block = getBlock("testBlock")
print("Original Block: ", block)
shiftedBlock = shiftRows(block)
print("Shifted rows: ",shiftedBlock)
````

**Task code written below (1):**

````python
# mixColumn() method here:
from copy import copy

# galoisMult(a,b) method given. See sample C and C++ 
# from the Wikipedia article for reference and comparison.
def galoisMult(a, b):
    p = 0
    hiBitSet = 0
    for _ in range(8):
        if b & 1 == 1: 
            p ^= a
        hiBitSet = a & 0x80
        a <<= 1
        if hiBitSet == 0x80:
            a ^= 0x1b 
        b >>= 1
    return p % 256

# Enter the statements within the method mixColumn correctly to finish this task.
def mixColumn(column):
    
    temp = copy(column) # Uncomment this line of code if you want to use the copy() method.
    
    # Enter statements for mixColumn here.
    
    column[0] = galoisMult(2,temp[0]) ^ galoisMult(3,temp[1]) ^ galoisMult(1,temp[2]) ^ galoisMult(1,temp[3]) # Perform Galois multiplication on each element of the first column
    column[1] = galoisMult(1,temp[0]) ^ galoisMult(2,temp[1]) ^ galoisMult(3,temp[2]) ^ galoisMult(1,temp[3]) # Perform Galois multiplication on each element of the second column
    column[2] = galoisMult(1,temp[0]) ^ galoisMult(1,temp[1]) ^ galoisMult(2,temp[2]) ^ galoisMult(3,temp[3]) # Perform Galois multiplication on each element of the third column
    column[3] = galoisMult(3,temp[0]) ^ galoisMult(1,temp[1]) ^ galoisMult(1,temp[2]) ^ galoisMult(2,temp[3]) # Perform Galois multiplication on each element of the fourth column
    
    return column
````

**Task code written below:**

````python
# keyScheduleCore() here:

def keyScheduleCore(word, itnum):
    
    rotWord = rotate(word,1) # Rotate the word by 1 bit
    subWord = []
    
    for i in range(len(rotWord)):
        
        subWord.append(Sbox[rotWord[i]]) # Substitute each element in the rotated word with the corresponding element in the Sbox
    
    subWord[0] = subWord[0] ^ Rcon[itnum] # XOR the first element in the substituted word with the corresponding element in the Rcon
    return subWord

word = [1,2,3,4]

newWord = keyScheduleCore(word,1)

print(word)

print(newWord)
    
````

**Task code written below:**

````python
# addRoundKey(block,roundKey) method here:

def addRoundKey(block, roundKey):
    
    for i in range(len(block)):
        
        block[i] = block[i] ^ roundKey[i] # XOR each element of the block with the corresponding element of the round key
    
    return block

testBlock = getBlock("testBlock")

key = getKey("testKey")


expandedKey = expandKey(key)

roundKey0 = createRoundKey(expandedKey,0)

addedRoundKeyToBlock = addRoundKey(testBlock,roundKey0) # Add the round key to the block

print(addedRoundKeyToBlock)
````

**Task code written below:**

````python
# encrypt(block, key) method here:

def encrypt(block, key):
# • Initialize:

    expandedKey = expandKey(key) # Get the expanded key


    for i in range(14): # Loop for 14 times
        if i == 0:
        #• 0-th round:
            roundKey0 = createRoundKey(expandedKey,0) # Get the round key for round 0
            modifiedBlock = addRoundKey(block, roundKey0) # Add the round key to the block
        
        if 1 <= i < 13:
        #• 1-th to 13:th round:
            roundKeyI = createRoundKey(expandedKey,i) # Get the round key for round i

            modifiedBlock = subBytes(modifiedBlock) # Substitute each element in the block with the corresponding element in the Sbox

            modifiedBlock = shiftRows(modifiedBlock) # Shift the rows of the block

            modifiedBlock = mixColumns(modifiedBlock) # Mix the columns of the block

            modifiedBlock = addRoundKey(modifiedBlock, roundKeyI) # Add the round key to the block
            
        if i == 13:
        #• 14:th round:
            roundKey14 = createRoundKey(expandedKey, 14) # Get the round key for round 14

            modifiedBlock = subBytes(modifiedBlock) # Substitute each element in the block with the corresponding element in the Sbox

            modifiedBlock = shiftRows(modifiedBlock) # Shift the rows of the block

            modifiedBlock = addRoundKey(modifiedBlock, roundKey14) # Add the round key to the block

    # • Return: – block
    return modifiedBlock

testBlock = getBlock("testBlock")
testKey = getKey("testKey")

print(testBlock)
encryptedBlock = encrypt(testBlock, testKey) # Encrypt the block
print(encryptedBlock)
````

**Task code written below:**

````python
# decrypt(block, key) method here:
def decrypt(block, key):
    
    #• Initialize:
    expandedKey = expandKey(key) # Get the expanded key
    
    for i in range(14,-1,-1): # Loop for 14 times, starting from 14 and ending at 0, with a step of -1
        if i == 14:
        #• 0-th round:
            roundKey14 = createRoundKey(expandedKey,14) # Get the round key for round 14
            modifiedBlock = addRoundKey(block, roundKey14) # Add the round key to the block
        
        if 1 <= i < 13:
        #• 1-th to 13:th round:
            roundKeyI = createRoundKey(expandedKey,i) # Get the round key for round i

            modifiedBlock = shiftRowsInv(modifiedBlock) # Inverse shift the rows of the block

            modifiedBlock = subBytesInv(modifiedBlock) # Substitute each element in the block with the corresponding element in the inverse Sbox

            modifiedBlock = addRoundKey(modifiedBlock, roundKeyI) # Add the round key to the block

            modifiedBlock = mixColumnsInv(modifiedBlock) # Inverse mix the columns of the block
            
        if i == 0:
        #• 14:th round:
            roundKey0 = createRoundKey(expandedKey, 0) # Get the round key for round 0

            modifiedBlock = shiftRowsInv(modifiedBlock) # Inverse shift the rows of the block

            modifiedBlock = subBytesInv(modifiedBlock) # Substitute each element in the block with the corresponding element in the inverse Sbox

            modifiedBlock = addRoundKey(modifiedBlock, roundKey0) # Add the round key to the block
    
    return modifiedBlock

testBlock = getBlock("testBlock")
testKey = getKey("testKey")

print(testBlock)
encryptedBlock = encrypt(testBlock, testKey) # Encrypt the block
print(encryptedBlock)
decryptedBlock = decrypt(encryptedBlock, testKey) # Decrypt the block
print(decryptedBlock)
````
